```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins  = pd.read_csv("penguins.csv")
penguins.head()
```

```python
penguins.info()
```

```python
overall_df = pd.DataFrame({
"Median" :penguins.median(numeric_only =True),
"Standard Deviation" :penguins.std(numeric_only=True),
"Mean": penguins.mean(numeric_only= True)
 })
overall_df
```

```python
penguins.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=penguins, x='body_mass_g', bins=30)
plt.title("Histogram of Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title("Boxplot of Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(data=penguins, x='bill_length_mm', bins=80)
plt.title("Histogram of Bill Length")
plt.xlabel("BILL LENGTH (MM)")
plt.ylabel("Frequency")
plt.show()

```

```python
import matplotlib.pyplot as plt

penguins['flipper_length_mm'].hist(bins=50, edgecolor='red')
plt.title("Histogram of flipper length")
plt.xlabel("flipper length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(data=penguins,x='sex',y='bill_length_mm')
plt.title("Boxplot of flipper length by bill length")
plt.xlabel("sex")
plt.ylabel("bill length (mm)")
plt.show()
```

```python
sns.boxplot(data=penguins,x='sex',y='flipper_length_mm')
plt.title("Boxplot of flipper length by bill length")
plt.xlabel("sex")
plt.ylabel("flipper length")
plt.show()
```

```python
sns.boxplot(data=penguins,x='sex',y='bill_length_mm')
plt.title("Boxplot of flipper length by bill length")
plt.xlabel("sex")
plt.ylabel("bill length (mm)")
plt.show()
```

```python
sns.scatterplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
plt.title("Flipper Length vs Bill Length")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Bill Length (mm)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(penguins.dropna())
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
sns.pairplot(penguins.dropna(), hue='species', palette='Set1')
plt.suptitle("Pairplot of Penguins by Species", y=1.02)
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.pairplot(penguins.dropna()[['bill_length_mm', 'flipper_length_mm', 'species']], hue='species', palette='Set1')
plt.suptitle("Pairplot of Bill Length and Flipper Length by Species", y=1.02)
plt.show()


```
